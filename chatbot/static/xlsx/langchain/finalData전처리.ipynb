```python
import pandas as pd
for i in range(4,8):
    문자열=str(i*500) + str((i+1)*500)
    # 두 개의 Excel 파일 읽기
    df_final = pd.read_excel('finaldata_merged.xlsx')
    df_100 = pd.read_excel('final_merged'+문자열+'.xlsx')

    # 병합 전 NaN 개수 확인
    nan_before = df_final['description_final'].isna().sum()
    print("병합 전 비어있는 description_final 수:", nan_before)

    # IMAGE_FILE 기준으로 임시 병합
    df_temp = pd.merge(
        df_final,
        df_100[['IMAGE_FILE', 'description_final']].rename(columns={'description_final': 'description_final_new'}),
        on='IMAGE_FILE',
        how='left'
    )

    # 기존 description_final이 NaN인 경우에만 새로운 값으로 채우기
    df_temp['description_final'] = df_temp['description_final'].fillna(df_temp['description_final_new'])

    # 임시 컬럼 제거
    df_final = df_temp.drop(columns=['description_final_new'])

    # 병합 후 NaN 개수 확인
    nan_after = df_final['description_final'].isna().sum()
    print("병합 후 비어있는 description_final 수:", nan_after)
    print(f"새로 채워진 데이터 수: {nan_before - nan_after}")

    # 결과 저장
    df_final.to_excel('finaldata_merged.xlsx', index=False)
    print("\n결과가 'finaldata_merged.xlsx'에 저장되었습니다.")

    # 샘플 데이터 출력
    print("\n처음 5개 행의 IMAGE_FILE과 description_final:")
    print(df_final[['IMAGE_FILE', 'description_final']].head())

    # 여전히 비어있는 IMAGE_FILE 확인
    still_empty = df_final[df_final['description_final'].isna()]['IMAGE_FILE']
    print(f"\n여전히 비어있는 IMAGE_FILE 수: {len(still_empty)}")
    if len(still_empty) > 0:
        print("첫 5개 비어있는 IMAGE_FILE:")
        print(still_empty.head())
```
